```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
unsample = pd.read_csv("AL_result_unsampling.txt", header=None)
```

```python
unsample.head()
```

```python
rand = pd.read_csv("AL_result_random.txt", header=None)
```

```python
rand.head()
```

```python
plt.plot(unsample[13], unsample[3], label = "unsampling")
plt.plot(rand[13], rand[3], label = "random")
plt.xlabel("epoch")
plt.ylabel("value")
plt.title("So sánh độ đo precision")
plt.legend()
plt.show()
```
